```python
carrier_bite = 0.1
army_bite = 0.05
death = 0.0001
carrier_kill = 0.05
army_kill = 0.8
training = 0.001

a_start = 0.05
c_start = 0.45
z_start = 0.5
d_start = 0.0

start_values = (a_start, c_start, z_start, d_start)

assert sum(start_values) == 1
```

```python
def diff_eqs(input, time):
    output = [0,0,0,0]
    output[0] = -(army_bite * input[0] * input[2]) - (death*input[0]) + (training * input[1]) # Army
    output[1] = -(carrier_bite * input[1] * input[2]) - (death*input[1]) - (training * input[1])#Carrier
    output[2] = (army_bite * input[0] * input[2]) + (carrier_bite * input[1] * input[2]) + (death*input[0]) + (death*input[1])\
                - (carrier_kill * input[1] * input[2]) - (army_kill * input[0] * input[2]) #Zombie
    output[3] = (carrier_kill * input[1] * input[2]) + (army_kill * input[0] * input[2]) #Dead
    return output
    assert sum(output) != 0, "Error"
    
```

```python
t_start = 0.0
t_end = 500
t_increment = 1
t_range = np.arange(t_start, t_end+t_increment, t_increment)
```

```python
result =  spi.odeint(diff_eqs, start_values, t_range)
print result

```
